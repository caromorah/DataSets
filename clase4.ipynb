```python
import pandas as pd
import numpy as np
```

```python
data = {'Company' : ['Ford', 'Ford', 'VW', 'Audi', 'Audi', 'Ford'],
        'Price': [20,22,30,34,25,28],
        'Weight': [55,60,70,28,30,80],
        'Length': [30, 35, 40, 40, 80, pd.NaT],
        'Made': ['China','Mexico','Mexico', 'China',np.nan, np.nan]}

df = pd.DataFrame(data = data)
df
```

```python
ndf = df.copy()
ndf
```

```python
ndf = ndf.dropna()
ndf
```
